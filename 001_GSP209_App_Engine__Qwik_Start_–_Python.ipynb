```python
"""Linia ta importuje klasę Flask z modułu Flask, który jest popularnym frameworkiem do tworzenia aplikacji webowych
w języku Python."""
from flask import Flask


"""Tworzy instancję klasy Flask i przypisuje ją do zmiennej app. Parametr __name__ wskazuje Flask,
gdzie znajduje się aplikacja."""
app = Flask(__name__)

"""Ta dekorator definiuje trasę dla endpointu '/' (czyli głównego adresu). Gdy użytkownik odwiedzi stronę główną,
zostanie wywołana funkcja hello(), która zwróci napis "Hello World!"."""
@app.route('/')
def hello():
    """Return a friendly HTTP greeting."""
    return 'Hello World!'

"""Ten warunek sprawdza, czy skrypt jest uruchamiany bezpośrednio (lokalnie, a nie jako importowany moduł).
Jeśli tak, to uruchamia lokalny serwer deweloperski Flask na porcie 8080.

W przypadku wdrożenia na Google App Engine, serwer internetowy, tak jak Gunicorn, zostanie użyty
do obsługi aplikacji."""
if __name__ == '__main__':

    app.run(host='127.0.0.1', port=8080, debug=True)
```
